```python
### REPORT ###

# Fonction pour vérifier si le report existe déjà dans notre BDD mySQL
def is_report_exist(code):
    
    #bon format pour la requête
    code = '"' + code + '"'
    
    # Connexion à la base de données
    conn = mysql.connector.connect(user=user, password=password, host=host, port=port)
    
    sql = """ SELECT * FROM projectfflogs.report WHERE id_report = """ + code + """; """
    
    df_result = pd.read_sql(sql, con=conn)

    if df_result.size == 0 :
        return False
    else :
        return True    
```

# Ajouts dans la base de données

Dans chacune des requêtes pour ajouter des données dans la base, on va mettre le curseur de connexion à la base de données.

Comme vu pour l'initialisation, l'idée est de garder une cohérence dans l'insertion des données et comme ces tables sont toutes liées les unes aux autres, il faut donc ouvrir un "transaction" au début de toutes les requêtes d'insertion qu'on veut effectuer et ene faire le commit() qu'à la toute fin.

Pour mieux situer où sont les erreurs lors de l'ajout, nous allons effectuer des try/except dans chaque fonction d'ajouts.

### Etape 1 : 'user'

```python
"""fonction qui ajoute un user dans la table user"""
def add_user(cursor, id_user, user_name):
    
    # vérifier s'il existe    
    exist = is_user_exist(id_user)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.user (id_user, user_name) VALUES (%s, %s)"""
        value = (id_user, user_name)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'user'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

### Etape 2 : 'static'

```python
"""fonction qui ajoute un static dans la table static"""
def add_static(cursor, id_static, static_name):
    
    # vérifier s'il existe    
    exist = is_static_exist(id_static)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.static (id_static, static_name) VALUES (%s, %s)"""
        value = (id_static, static_name)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'static'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

### Etape 3: 'report'

```python
""" fonction qui ajoute un report dans la table report"""
def add_report(cursor, id_report, report_date, fk_owner, fk_id_static):
    
    # vérifier s'il existe    
    exist = is_report_exist(id_report)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.report (id_report, report_date, fk_owner, fk_id_static) VALUES (%s, %s, %s, %s)"""
        value = (id_report, report_date, fk_owner, fk_id_static)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'report'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

### Etape 4 : 'try'

```python
""" fonction qui ajoute un try dans la table try"""
def add_try(cursor, id_try, fk_id_report, fk_id_encounter, boss_percentage, encounter_percentage, is_clear, start_time, end_time):
    
    # vérifier si l'utilisateur existe    
    exist = is_try_exist(id_try, fk_id_report)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.try (id_try, fk_id_report, fk_id_encounter, \
        boss_percentage, encounter_percentage, is_clear, start_time, end_time) \
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s);"""
        value = (id_try, fk_id_report, fk_id_encounter, boss_percentage, encounter_percentage, is_clear, start_time, end_time)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'try'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

### Etape 5 : 'character'

```python
""" fonction qui ajoute un character dans la table character"""
def add_character(cursor, id_character, character_name, server, region, slug):
    
    # vérifier si l'utilisateur existe    
    exist = is_character_exist(id_character)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.character (id_character, character_name, server, region, slug) \
        VALUES (%s, %s, %s, %s, %s);"""
        value = (id_character, character_name, server, region, slug)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'character'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

### Etape 6 : 'teampertry'

```python
"""fonction qui ajoute un teampertry dans la table character"""
def add_teampertry(cursor, fk_id_try, fk_id_report, fk_id_character, fk_id_job):
    
    # vérifier si l'utilisateur existe    
    exist = is_teampertry_exist(fk_id_try, fk_id_report, fk_id_character, fk_id_job)
    
    # s'il n'existe pas on l'ajoute
    if not exist :
        
        sql = """INSERT INTO projectfflogs.teampertry (fk_id_try, fk_id_report, fk_id_character, fk_id_job) \
        VALUES (%s, %s, %s, %s);"""
        value = (fk_id_try, fk_id_report, fk_id_character, fk_id_job)
        
        try :
            cursor.execute(sql, value)
        except Exception as error:
            print("Erreur lors de l'insertion dans la table 'teampertry'.", error)
    
    ##### QUE CA MARCHE OUPAS ON RENVOIE LE CURSEUR EN l'ETAT ????
    return cursor
```

# Fonction pour récupérer des données de l'API FFlogs

(Attention aux nombres d'appel par heure, les points sont comptés !)

```python
"""
Fonction envoie la query donné en paramètre à l'apifflogs V2 et renvoie un dictionnaire json
"""
def query_apifflogsV2(query):    
    response = requests.get(url, auth=BearerAuth(token), json={'query': query})    
    return response.json()
```

# Fonction pour ajouter un rapport dans la base de données

```python
#fonction pour ajouter les information d'un rapport de données

def add_information_from_a_report(code):
    
    
    #initialisation de la variable list_encounter qui nous sera fort utile par la suite
    list_encounter = encounter_to_list()
    print("List_encounter au tout début : ", list_encounter)
     
    ### Connexion à notre base de données ###
    # Connexion à la base de données
    conn = mysql.connector.connect(user=user, password=password,
                                  host=host, port=port)
    
    #creation du cursor
    cursor = conn.cursor()
    
    #DEBUT TRANSACTION       
    #ETAPE 0 : Est-ce que le rapport existe dans notre base de donnée MySQL ?
    report_exist = is_report_exist(code)
    if report_exist :
        return "le rapport existe déjà" 
    
    #ENSUITE, on appelle l'api avec le code donné en paramètre 'code' avec la query suivante
    query = """{
      reportData { 
            report(code: """+str('"' + code + '"')+""") {
                startTime, code
                fights{
                    id, 
                    name,
                    difficulty,
                    kill, 
                    startTime, 
                    endTime, 
                    bossPercentage, 
                    fightPercentage, 
                    encounterID, 
                    friendlyPlayers, 
                    size},
                owner{id,name},
                guild{name,id},
                masterData{
                    actors{
                        id,
                        name, 
                        type, 
                        subType,
                        server
                    }
                },
                rankedCharacters{id,name,server{name,region{name,slug}}}
            }
        }
        }
        """
    json_data = query_apifflogsV2(query)
    
    #pprint(json_data)
    
    id_report = json_data["data"]["reportData"]["report"]["code"]
    report_date = float(json_data["data"]["reportData"]["report"]["startTime"]) #à convertir en datetime
    fk_owner = json_data["data"]["reportData"]["report"]["owner"]["id"]
    fk_id_static = json_data["data"]["reportData"]["report"]["guild"]["id"]
    report_date = dt.datetime.fromtimestamp(report_date/1000)
    
    try :
        
        #ETAPE 1 : ajout du owner dans la table user (s'il existe pas déjà)
        id_user = fk_owner
        user_name = json_data["data"]["reportData"]["report"]["owner"]["name"]
        cursor = add_user(cursor, id_user, user_name)
        
        #ETAPE 2 : ajouter le static dans la table static (s'il existe pas déjà)
        id_static = fk_id_static
        static_name = json_data["data"]["reportData"]["report"]["guild"]["name"]
        cursor = add_static(cursor, id_static, static_name)
        
        #ETAPE 3: ajouter dans la table report (enfin)
        cursor = add_report(cursor, id_report, report_date, fk_owner, fk_id_static)
        
        #ETAPE 4 : ajouter dans la table try
        # on récupère la liste des fights du report en question
        fights = json_data["data"]["reportData"]["report"]["fights"]
        #liste pour savoir quels trys sont gardés
        validated_fights = []
        fk_id_report = id_report
        for fight in fights :
            id_try = fight["id"]
            fk_id_encounter = fight["encounterID"]
            boss_percentage = fight["bossPercentage"]
            encounter_percentage = fight["fightPercentage"]
            is_clear = int(fight["kill"]) if fight["kill"] is not None else 0
            start_time = fight["startTime"]
            end_time = fight["endTime"]
            print("fk_id_encounter :", fk_id_encounter)
            print("list_encounter :",list_encounter)
            if fk_id_encounter in list_encounter:
                print("avant cursor add_try")
                cursor = add_try(cursor, id_try, fk_id_report, fk_id_encounter, boss_percentage, encounter_percentage, is_clear, start_time, end_time)
                validated_fights.append(id_try)

        #ETAPE 5 : ajouter dans la table character
        # on récupère tous les personnages présent dans le rapport
        characters =  json_data["data"]["reportData"]["report"]["masterData"]["actors"]
        df_characters = pd.DataFrame(characters, columns=['id', 'name', 'server', 'subType', 'type'])
        df_characters = df_characters.rename(columns={'id': 'id_actor', 'subType': 'job'}).drop("type", axis=1)
        cond = ~(df_characters["server"].isnull())
        #on a la liste de tous les personnages qui apparaissent dans le rapport
        df_characters = df_characters[cond]        
        # pour le moment on va se contenter de regarder la rankedCharacters
        ranked_chars = json_data["data"]["reportData"]["report"]["rankedCharacters"]
        rename = {'server.name': 'server', 'server.region.name': 'region', 'server.region.slug': 'slug'}
        df_ranked_chars = pd.json_normalize(ranked_chars).rename(columns=rename)
        # merge pour recup toutes les informations des deux df
        df_char = pd.merge(df_characters, df_ranked_chars, on=["name","server"], how="inner")
        df_char["name"] = df_char["name"].apply(str)
        for c in df_char.itertuples() :
            cursor = add_character(cursor, c.id, c.name, c.server, c.region, c.slug)
        
        
        
        #ETAPE 6 : ajouter dans la table teampertry
        df_jobs = job_to_df()
        df_jobs['job'] = df_jobs['job_name'].str.replace(' ', '')
        df_jobs = df_jobs.drop('job_name',axis=1)
        
        # on va commercer par ajouter id_job au df_char
        df_char_job = pd.merge(df_char, df_jobs, on='job', how="inner")
        print(df_char_job.head())
        
        for fight in fights :
            print("boucle fight de fights.")
            id_try = fight["id"]
            print("id_try", id_try)
            print("validated_fights", validated_fights)
            if id_try in validated_fights :
                print("if validated_fights, id_try=",id_try)
                #on récupère la liste des friendlyplayers
                friendlyplayers =  fight["friendlyPlayers"] # liste d'entier
                #pour récup id_character et id_job
                for friend in friendlyplayers :
                    print("boucle avant teampertry friendlyplayers")
                    cond = df_char_job["id_actor"] == int(friend)
                    df_player = df_char_job[cond]
                    if not df_player.empty:
                        print("boucle teampertry")
                        id_character = int(df_player["id"].values[0])
                        id_job = int(df_player["id_job"].values[0])
                        cursor = add_teampertry(cursor, id_try, id_report, id_character, id_job)

        # on commit le tout !
        conn.commit()
    except Exception as error:
        print("Erreur pendant l'ajout des informations du rapport, aucune table n'a été modifiée.", error) 
    
```
