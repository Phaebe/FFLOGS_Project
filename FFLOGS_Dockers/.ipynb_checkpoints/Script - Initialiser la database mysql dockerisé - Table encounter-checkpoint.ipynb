# Remplir la table 'job'

```python
# REMPLIR la table 'job'

#Requête à l'api fflogs v2 pour récupérer la liste des  jobs
query = """query {
    gameData{
            classes{
                specs{
                    id, name
                }
            }
        }
}

"""

#on récupère une liste de jobs à charger dans la BDD
r = requests.get(url, auth=BearerAuth(token), json={'query': query}) 
r.json()
json = r.json()

# On en tire une liste de jobs
spec = json["data"]["gameData"]["classes"]
jobs = spec[0]["specs"]


# Pour chaque job de la liste
for job in jobs:
    job_name = job["name"]
    job_id = job["id"]
    
    #ON VA DETERMINER IL EST DE QUEL TYPE
    # Tank are jobs speciliazed in defense ("Dark Knight", "Paladin", "Warrior", "Gunbreaker")
    # Heal group all jobs speciliazed in healling ("Astrologian", "Scholar", "White Mage", "Sage")
    # Melee group all melee jobs ("Dragoon", "Monk", "Ninja", "Samurai", "Reaper") 
    # Range group all physical range jobs ("Bard", "Machinist", "Dancer")
    # Caster group all magical range attacks' jobs ("Black Mage", "Summoner", "Red Mage")
    if job_name in ["Dark Knight", "Paladin", "Warrior", "Gunbreaker"]:
        jobtype_id = 0
    elif job_name in ["Astrologian", "Scholar", "White Mage", "Sage"]:
        jobtype_id = 1
    elif job_name in ["Dragoon", "Monk", "Ninja", "Samurai", "Reaper"]:
        jobtype_id = 2
    elif job_name in ["Bard", "Machinist", "Dancer"]:
        jobtype_id = 3
    elif job_name in ["Black Mage", "Summoner", "Red Mage"]:
        jobtype_id = 4
    else:
        jobtype_id = 99
    #print("jobtype_id = ", jobtype_id, ":", job_name)
    
    # MTN on a le job_name, id_jobtype, il n'y a plus qu'à rajouter le tout dans la table
    # en incrémentant un l'id de chaque job
    element = (job_id, job_name, jobtype_id)
    #print(element)
    cursor.execute("""INSERT INTO projectfflogs.job (id_job, job_name, fk_id_job_type) VALUES(%s, %s, %s)""", element)
    
#on commit pour que les changements soient effectifs dnas la base
conn.commit()
```

```python
# Création d'un dataframe vide pour stocker les codes des rapports
df_total = pd.DataFrame()

# LA Query général avec toutes les données de base ! - sera complété par d'autres query pour certaines tables

code = '"TfYqLdD2X7gb4pVx"'
query = """{
      reportData { 
            report(code: """+str(code)+""") {
                startTime, code
                fights{
                    id, 
                    name,
                    difficulty,
                    kill, 
                    startTime, 
                    endTime, 
                    bossPercentage, 
                    fightPercentage, 
                    encounterID, 
                    friendlyPlayers, 
                    size},
                owner{id,name},
                guild{name,id},
                masterData{
                    actors{
                        id,
                        name, 
                        type, 
                        subType,
                        server
                    }
                },
                rankedCharacters{id,name,server{name,region{name,slug}}}
            }
        }
}
"""

response = requests.get(url, auth=BearerAuth(token), json={'query': query})
json_data = response.json()

id_report = json_data["data"]["reportData"]["report"]["code"]
report_date = float(json_data["data"]["reportData"]["report"]["startTime"]) #à convertir en datetime
fk_owner = json_data["data"]["reportData"]["report"]["owner"]["id"]
fk_id_static = json_data["data"]["reportData"]["report"]["guild"]["id"]

report_date = dt.datetime.fromtimestamp(report_date/1000) 

print(report_date)
```

```python
# on récupère tous les personnages présent dans le rapport
characters =  json_data["data"]["reportData"]["report"]["masterData"]["actors"]

df_characters = pd.DataFrame(characters, columns=['id', 'name', 'server', 'subType', 'type'])
df_characters = df_characters.rename(columns={'id': 'id_actor', 'subType': 'job'}).drop("type", axis=1)

cond = ~(df_characters["server"].isnull())

#on a la liste de tous les personnages qui apparaissent dans le rapport
df_characters = df_characters[cond]

df_characters
```

##### Remplir la table 'teampertry'

- fk_id_try : int(11)
- fk_id_report : varchar(255)
- fk_id_character : int(11)
- fk_id_job : int(11)

```python
# on recup la liste des jobs 
query_job = """query {
    gameData{
            classes{
                specs{
                    id, name
                }
            }
        }
}

"""

#on récupère une liste de jobs à charger dans la BDD
r = requests.get(url, auth=BearerAuth(token), json={'query': query_job})
json = r.json()
jobs = json["data"]["gameData"]["classes"][0]["specs"]
df_jobs = pd.json_normalize(jobs)
df_jobs['name_subtype'] = df_jobs['name'].str.replace(' ', '')
rename= {'name_subtype':'job', 'id':'id_job'}
df_jobs = df_jobs.rename(columns=rename).drop('name',axis=1)
df_jobs
```

```python
# on va commercer par ajouter id_job au df_char
df_char_job = pd.merge(df_char, df_jobs, on='job', how="inner")
df_char_job
```

```python
# il faut pour un fight donné (avec l'id_report et id_try) du report, mettre l'id_character et son job
# dans la liste des fights, il y a la donnée "friendlyPlayers" 
# qui listes tous les personnages acteurs présents dans le fight
# on va le lié à l'id_actor du df_character qui nous permet de retrouver l'information du job du perso sur ce fight

# on avait récupéré la liste de tous les fights dans la variable fights 
#pprint(fights["id"])

sql = """INSERT INTO projectfflogs.teampertry (fk_id_try, fk_id_report, fk_id_character, fk_id_job) values(%s, %s, %s, %s);"""

for fight in fights :
      
    id_try = fight["id"]
    
    #on récupère la liste des friendlyplayers
    friendlyplayers =  fight["friendlyPlayers"] # liste d'entier
    
    #pour récup id_character et id_job
    for friend in friendlyplayers :
        
        cond = df_char_job["id_actor"] == int(friend)
        df_player = df_char_job[cond]
        
        if not df_player.empty:
            id_character = int(df_player["id"].values[0])
            id_job = int(df_player["id_job"].values[0])

            element = (id_try, id_report, id_character, id_job)
            print(element)
            
            ##### Rajouter code BDD #####
            cursor.execute(sql, element)  
```
